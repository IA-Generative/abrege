```python
%load_ext autoreload
%autoreload 2
from py_pdf_parser.loaders import load_file

document = load_file("1805.07799v1.pdf")
text2 = ""
for element in document.elements:
    text2 += element.text()

from extractive_summary import EmbeddingModel
embedding_model = EmbeddingModel.from_hugging_hub_sentence_transformer("dangvantuan/sentence-camembert-base")
```

```python
from summary_chain import summarize_chain_builder

custom_chain = summarize_chain_builder(method="text_rank", embedding_model=embedding_model)
res = custom_chain.invoke(text2)
res
```

```python
custom_chain_map_reduce = summarize_chain_builder(method="map_reduce")
res2 = custom_chain_map_reduce.invoke(text2)
```

```python
from bart_score import BARTScorer

bart_scorer = BARTScorer()
```

```python
score = bart_scorer.score(text2, res)
```

```python
score2 = bart_scorer.score(text2, res2)
```

```python
print(sum(score))
print(sum(score2))
```
